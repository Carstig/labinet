```python
import os

CWD_PATH = os.path.join(os.getcwd(),'..') 
IMAGE_PATH = os.path.join(CWD_PATH,'images')
```

```python
files = os.listdir(IMAGE_PATH)
xml_files = []
jpg_files = []
for file in files:
    if file.endswith('xml'):
        xml_files.append(file[:-4])
    elif file.endswith('jpg'):
        jpg_files.append(file)

jpg_without_xml = []        
        
for jpg in jpg_files:
    fname = jpg[:-4]
    if fname in xml_files:
        pass
    else:
        jpg_without_xml.append(jpg)
        

```

```python
print(jpg_without_xml)
```
